```python
import codecs
import pandas as pd
```

```python
output = pd.read_csv('final.txt', header = None, names = ['name','parent','rating','count'], sep='\t')
kanmusu_dict = pd.read_csv('kantai_chara_dict_kanmusu_only.txt', header = None, names = ['name','tag','type'], sep = '\t')
shinkai_dict = pd.read_csv('kantai_chara_dict_shinkai_only.txt', header = None, names = ['name','tag','type'], sep = '\t')

kanmusu_dict = kanmusu_dict.drop(columns=['tag'])
shinkai_dict = shinkai_dict.drop(columns=['tag'])

print(output.head())
print(kanmusu_dict.head())
print(shinkai_dict.head())
```

```python
output = pd.merge(output, kanmusu_dict, on = 'name', how = 'left')
output = pd.merge(output, shinkai_dict, on = 'name', how = 'left')
output.head()
```

```python
def label_kanmusu_type (row):
    if pd.notnull(row['type_x']):
        return '舰娘'
    elif pd.notnull(row['type_y']):
        return '深海'
    return '其他'

output['kanmusu_shinkai_flag'] = output.apply(lambda row: label_kanmusu_type(row), axis = 1)
output
```

```python
def label_type (row):
    if pd.notnull(row['type_x']):
        return row['type_x']
    elif pd.notnull(row['type_y']):
        return row['type_y']
    return '不适用'

output['type'] = output.apply(lambda row: label_type(row), axis = 1)
output
```

```python
output.loc[output['parent']=='all', 'parent']='全部'
output.loc[output['parent']=='parent:none', 'parent']='无子项'
output.loc[output['rating']=='all', 'rating']='全部'
output.loc[output['type']=='Other', 'type']='其他'
output.loc[output['type']=='Installation', 'type']='路基'

output
```

```python
output = output.drop(columns=['type_x','type_y'])
```
