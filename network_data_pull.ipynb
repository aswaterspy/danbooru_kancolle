```python
dict_file = codecs.open('kantai_chara_dict_kanmusu_pure.txt', 'r', 'utf-8')
chara_dict = {}

for line in dict_file:
    name_pair = (line[:-1]).split('\t')
    chara_dict[name_pair[0]] = name_pair[1]

print(chara_dict['叢雲'])
print(chara_dict['島風'])
# chara_dict
```

```python
def count_cp(chara_dict):
    
    output_dict = {}
    comp_dict = chara_dict.copy()
    
    for i in chara_dict:
        del comp_dict[i]
        for j in comp_dict:
            
            #build query url
            chara_url = 'https://danbooru.donmai.us/counts/posts?tags='
            if i == '島風' or j == '島風':
                chara_url = chara_url + '+' + chara_dict[i] + '+' + comp_dict[j] + '+-shimakaze_%28kantai_collection%29_%28cosplay%29'
            else:
                chara_url = chara_url + '+' + chara_dict[i] + '+' + comp_dict[j]

            #get query url and count
            chara_count_page = session_requests.get(chara_url)
            chara_count_query = pq(chara_count_page.text)

            for tag_div in chara_count_query('div'):
                div_query = pq(tag_div)
                div_id = div_query.attr('id')
                if div_id == 'a-posts':
                    count = div_query.remove('a').text()
                    count = re.findall(r'\d+', count)[0]

            output_dict_key = (i,j)
            output_dict[output_dict_key] = count
    
            final_output_file = codecs.open('cp_output.txt', 'a', 'utf-8')
            final_output_file.write('{}\t{}\t{}\n'.format(i,j,count))
            final_output_file.close()
            
            print(i,j,count)
            # Waits before sending next request
            time.sleep(1 + randint(0,5))
        
    return output_dict
```

```python
cp_output = count_cp(chara_dict)
```
